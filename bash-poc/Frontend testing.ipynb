# SASSファイルの追加

```bash
echo """
.c-sidebar {
  margin-right: -21px; /* 20px padding + 1px border */
  margin-bottom: 20px;
  margin-left: -20px;

  .c-sidebar-title {
    font-weight: 300;
    text-transform: uppercase;
    color: #888;
    padding-left: 20px;
    line-height: 36px;
  }

  .c-sidebar-menu {

    a {
      display: table-cell;
      word-break: break-all;
      vertical-align: middle;
      padding-left: 40px;
      padding-right: 10px;
      height: 34px;
      width: 16.66666667%;
      position: relative;
      border-top: 1px solid transparent;
      border-bottom: 1px solid transparent !important;

      &:link,
      &:visited {
        color: #555;
        border-bottom: 0;
        text-decoration: none;
      }
    }
    
    &.current > a {
      font-weight: 700;
      background-color: #fff;
      border-top: 1px solid #d2d2d2;
      border-bottom: 1px solid #d2d2d2 !important;

      &:before {
        content: '';
        width: 10px;
        // background-color: #348899;
        background-color: #f0ad4e;
        position: absolute;
        top: 0;
        bottom: 0;
        left: -1px;
      }
    }

    &:not(.current) > a:hover {
      color: #3785a9;
      border-bottom: 0;
      text-decoration: none;
    }

  }
}
""" >> ./sass/test.scss
```

```bash
echo """
'use strinct';
var gulp = require('gulp');
var sass = require('gulp-sass');

gulp.task('sass', function(){
  return gulp.src('./sass/*.scss').pipe(sass().on('error', sass.logError)).pipe(gulp.dest('./css'));
});
""" >> gulpfile.js
```
